# Data Cleaning

- Remove text values and convert to numbers


##### Changing Datatypes

```python
df['fuel economy'] = pd.to_numeric(                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    
    df['fuel economy']
    .astype(str)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    
    .str.replace(r'[^0-9.-]','',regex=True)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    
    .replace('',None), 
    errors='coerce'
)
```

```python
df
```

```python
df['torque'] = pd.to_numeric(
    df['torque']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
```

```python
df
```

```python
df['max_power'] = pd.to_numeric(
    df['max_power']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
```

```python
df.info()
```

```python
df['engine'] = pd.to_numeric(
    df['engine']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
```

```python
df
```

### handling missing values

```python
df.isnull().sum()
```

```python
df.dropna()
```

- Decision
  >Through dropping rows,222 vehicles is dropping.it will be a huge loss for company so we will fixing it through some methods.

1. engine rows filling through median method

```python
df['engine']=df['engine'].fillna(df['engine'].median())
```

2. seat rows filling through 

```python
df['seats']=df['seats'].fillna(5)
```

3. fuel economy

```python
df['fuel economy']=df['fuel economy'].fillna(df['fuel economy'].median())
```

4. Max power row filling through median

```python
df['max_power']=df['max_power'].fillna(df['max_power'].median())
```
